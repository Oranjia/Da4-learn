```python
"""
给定关键字,百度搜索,把对应的html文件给获取下来
"""
from urllib import request
from urllib import parse

#1、拼接URL地址
word=input("请输入百度搜索的关键字")
params=parse.urlencode({"wd":word})
url="http://www.baidu.com/s?{}".format(params)
headers={
   "User-Agent":"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/535.1 (KHTML, like Gecko) Chrome/14.0.835.163 Safari/535.1"
}

#2、发送请求
req=request.Request(url=url,headers=headers)
res=request.urlopen(req)
html=res.read().decode()

#3、保存到本地文件
filename=word+".html"
with open(filename,"w",encoding="utf-8") as f:
    f.write(html)
```

```python
"""
正则表达式
"""
import re
#用re.S .就可以匹配换行符了
html="""
<div><p>wxl</p></div>
<div><p>goushi</p></div>
"""

# 贪婪匹配
pattern=re.compile("<div><p>.*</p></div>",re.S)
r_list=pattern.findall(html)
print(r_list)

# 非贪婪匹配
pattern=re.compile("<div><p>.*?</p></div>",re.S)
r_list=pattern.findall(html)
print(r_list)


"""
正者则表达式分组示例
"""
html="A B C D"
pattern=re.compile("\w+\s+\w+") #\s+是空白
r_list=pattern.findall(html)
print(r_list)

pattern=re.compile("(\w+)\s+\w+") #\s+是空白
#第一步：['A B', 'C D'],第二步：单独搞 A C
r_list=pattern.findall(html)
print(r_list)

pattern=re.compile("(\w+)\s+(\w+)") #\s+是空白
##第一步：括号先不看，['A B', 'C D'],第二步：多个分组，将数据放到元组中作为不同的元素 
r_list=pattern.findall(html)
print(r_list)

```

```python
# 好好整理一下  使用writerows   Mongo数据库
from urllib import request,parse
import time
import random
import pymongo
import re
class DoubanSpider:
    def __init__(self):
        """定义常用的变量"""
        self.url="https://movie.douban.com/chart"
        self.headers={
            "User-Agent":"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/535.1 (KHTML, like Gecko) Chrome/14.0.835.163 Safari/535.1"
        }
        self.conn=pymongo.MongoClient(host="localhost",port=27017)
        self.db=self.conn["doubandb"]
        self.myset=self.db["doubanset"]

    def get_html(self,url):
        """获取响应内容的函数"""
        req=request.Request(url=url,headers=self.headers)
        res=request.urlopen(req)
        html=res.read().decode()
        #直接调用解析函数
        self.parse_html(html)

    def parse_html(self,html):
        """解析提取数据的函数"""
        regex='<div class="pl2">.*?<span style="font-size:13px;">(.*?)</span>.*?<p class="pl">(.*?)</p>.*?<span class="rating_nums">(.*?)</span>.*?<span class="pl">(.*?)</span>'
        pattern =re.compile(regex,re.S)
        r_list=pattern.findall(html)
        #直接调用数据处理函数
        self.save_html(r_list)

    def save_html(self,r_list):
        """数据处理的函数"""
        
        for i in r_list:
            item={} #只存这三个，然后让mangodb去自动的给id
            item["name"]=i[0].strip()
            item["info"]=i[1].strip()
            item["score"]=i[2].strip()
            item["people"]=i[3].strip()
            self.myset.insert_one(item) #一行一行插入
            print(item)

    def run(self):
        """程序入口函数"""
        url=self.url
        self.get_html(url)



if __name__=="__main__":
    spider=DoubanSpider()
    spider.run()
```
