```python
import requests
import json
import time
import random
from fake_useragent import UserAgent
import re
from urllib import parse

class DoubanSpider:
    def __init__(self):
        self.url='https://movie.douban.com/j/new_search_subjects?sort=U&range=0,10&tags=&start={}&genres={}'
        self.f=open("doubandianying.json","w",encoding="utf-8")
        self.item_list=[] 

    def get_agent(self):
        """获取随机USER-AGENT"""
        return UserAgent().random   

    def get_html(self,url):
        """获取html"""
        html=requests.get(url=url,headers={"User-Agent":self.get_agent()}).text
        return html
    
    def parse_html(self,url):
        html=json.loads(self.get_html(url=url)) #json loads 可以把json格式的字符串转为python数据类型
        html=html["data"]
        # print(html)
        item={}
        for one_film in html:
            # print(one_film)
            item["name"]=one_film["title"]
            item["score"]=one_film["rate"] #分数
            # one_film["rank"]=one_film["rate"]  #排名
            # item["time"]=one_film[""]   #上映时间
            print(item)
            self.item_list.append(item)
    def run(self):
        type_dict=self.get_type_dict()
        menu=''
        for t in type_dict:
            menu = menu +t +" "
        print(menu)
        choice=input("请输入你要抓取的类别：")
        params=parse.quote(choice)
        # typ=type_dict[choice]
        for start in range(0,100,20):
            url=self.url.format(start,params)
            self.parse_html(url=url)
            time.sleep(random.randint(1,2))
        #都抓完了
        json.dump(self.item_list,self.f,ensure_ascii=False)
        self.f.close()

    def get_type_dict(self):
        """获取所有类别的字典"""
        url='https://movie.douban.com/chart'
        html=self.get_html(url=url)
        regex='<span><a href=".*?type_name=(.*?)&amp;type=(.*?).*?'
        pattern=re.compile(regex,re.S)
        r_list=pattern.findall(html) #[("剧情"，"11"),]
        # print(r_list)
        type_dict={}
        for r in r_list:
            type_dict[r[0]]=r[1]
        
        return type_dict



if __name__=="__main__":
    spider=DoubanSpider()
    spider.run()




            

```
