```python

# https://tieba.baidu.com/f?kw=q宠大乐斗&ie=utf-8&pn=0|50|100|150

# 一级页面提取帖子的链接，二级页面提取链接中的图片

#href  ：  //div[@class="t_con cleafix"]/div/div/div/a/@href

#图片 //div[@class="d_post_content j_d_post_content "]//img[@class="BDE_Image"]/@src
```

```python
"""
抓取指定贴吧中的图片   视频的话也是一个道理
"""
import requests
from lxml import etree
import time
import random
from urllib import parse

class TiebaImageSpider:
    def __init__(self):
        self.url='https://tieba.baidu.com/f?kw={}&pn={}'
        #使用ie的
        self.headers={"User-Agent":'Mozilla/5.0 (compatible; MSIE 9.0; Windows NT 6.1; Win64; x64; Trident/5.0; .NET CLR 2.0.50727; SLCC2; .NET CLR 3.5.30729; .NET CLR 3.0.30729; Media Center PC 6.0; InfoPath.3; .NET4.0C; Tablet PC 2.0; .NET4.0E)'}
    
    def get_html(self,url):
        html=requests.get(url=url,headers=self.headers).text
        # print(html)
        return html
    
    def xpath_func(self,html,xpath_bds):
        p=etree.HTML(html)  #创建一个解析对象
        r_list=p.xpath(xpath_bds)
        return r_list

    def get_image(self,href):
        """把一个帖子中保存到本地"""
        two_url="http://tieba.baidu.com"+href
        two_html=self.get_html(url=two_url)
        two_xpath='//img[@class="BDE_Image"]/@src'
        img_list=self.xpath_func(html=two_html,xpath_bds=two_xpath)#每一个图片的链接
        # print(img_list)
        for img in img_list:
            html=requests.get(url=img,headers=self.headers).content  #content是byte类型
            filename=img[-10:]+".jpg" #倒数10
            with open (filename,"wb") as f:
                f.write(html)
            print(filename,"下载成功")
            time.sleep(random.randint(0,1))

    def parse_html(self,one_url):
        """一级页面：提取帖子链接，依次向每个链接发送请求，最终下载图片"""
        one_html=self.get_html(url=one_url)
        one_xpath='//div[@class="t_con cleafix"]/div/div/div/a/@href'
        href_list=self.xpath_func(html=one_html,xpath_bds=one_xpath)
        for href in href_list:
            #拿到一个帖子的链接后，把这个帖子中的所有图片保存到本地
            self.get_image(href)
    
    def run(self):
        name=input("请输入贴吧名字：")
        start=int(input("请输入起始页："))
        end=int(input("请输入结束页："))
        name=parse.quote(name)
        for i in range(start,end+1):
            pn=(i-1)*50
            one_url=self.url.format(name,pn)
            self.parse_html(one_url)

if __name__ =="__main__":
    spider=TiebaImageSpider()
    spider.run()


            






```
